```python
import pandas as pd
import matplotlib.pyplot as plt
import sys
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
```

```python
df=pd.read_csv("bed_vacancy_details_under_covid_health_centres_and_covid_district_hospitals_in_tamil_nadu_as_on_28_09_2021.csv")
```

```python
figure(figsize=(10, 8), dpi=80)
plt.scatter(df['Total Vacant beds under CHC and CDH as on 28.09.2021'],df['Vacancy of beds with O2 supply under CHC and CDH as on 28.09.2021'])
plt.xlabel("Total Vacant beds under CHC and CDH as on 28.09.2021")
plt.ylabel("Vacancy of beds with O2 supply under CHC and CDH as on 28.09.2021")
```

```python
figure(figsize=(8, 6), dpi=80)
plt.xticks(rotation='vertical')
plt.ylabel("Total Vacant Beds")
plt.bar(df['Name of the District'],df['Vacancy of beds with O2 supply under CHC and CDH as on 28.09.2021'])
```

```python
figure(figsize=(8, 6), dpi=80)
plt.boxplot([df['Total Vacant beds under CHC and CDH as on 28.09.2021'],df['Vacancy of beds with O2 supply under CHC and CDH as on 28.09.2021']],labels=["Total Vacant Beds","Vacant beds with 02 Supply"])
plt.ylabel("Total Number of Beds")
```
